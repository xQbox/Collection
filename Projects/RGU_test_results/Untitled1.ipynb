```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("2.csv")
df.head(0)
```

```python
del df['NN']
```

```python
df
df = df.rename(columns={'Data_chem': 'data'})
df = df.rename(columns={'data_rus': 'data1'})
df
```

```python
first = df[['data','Chemistry_res']]
second = df[['data1', 'Russian_res']]
```

```python
first = first.sort_values(by=['data'])
first = first.reset_index(drop=True)
first = first.drop(df.index[-4:])
first
```

```python
second = second.sort_values(by=['data1'])
second = second.reset_index(drop=True)
second
```

```python

second = second.rename(columns={'data1': 'data'})
```

```python
df = pd.merge(first, second, on='data')
# result = pd.merge(df1, df2, on='key', how='inner')
```

```python
df
```

```python
def can_convert_to_int(value):
    try:
        int(value)
        return False
    except ValueError:
        return True

def replace_strings_with_number(x):

    if can_convert_to_int(x):
        return 0
    return x

df['Chemistry_res'] = df['Chemistry_res'].apply(replace_strings_with_number)
df['Russian_res'] = df['Russian_res'].apply(replace_strings_with_number)
```

```python
df
```

```python

def update_values(row):
    if row['Chemistry_res'] == 0 or row['Russian_res'] == 0:
        return 0
    return int(row['Chemistry_res']) + int(row['Russian_res'])

df['Total'] = df.apply(update_values, axis=1)
```

```python
df
```

```python
df = df.sort_values(by='Total', ascending=False)
```

```python
df
```
